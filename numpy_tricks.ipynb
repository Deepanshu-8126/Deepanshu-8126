```python
# SORTING (sort) np.sort
# Return a sorted copy of an array
import numpy as np
a=np.random.randint(1,100,15)
b=np.random.randint(1,100,24).reshape(6,4)
print(a)
print(b)

# sort
print(np.sort(a)) # aascending order sorting

```

```python
# descending order sorting
print(np.sort(a)[::-1])
```

```python
# row wise and column wise sorting using axis
print(b)
np.sort(b,axis=0)
np.sort(b,axis=1) # column wise
```

```python
# 2- function np.append
print(a)
print(np.append(a,200)) # append add on the last


# for 2d array
print(np.append(b,np.ones((b.shape[0],1)),axis=1))  # row

# column
print(np.append(b,np.random.random((b.shape[0],1)),axis=1))
```

```python
# Concatenate
# in 2d arrays use most
c=np.arange(6).reshape(2,3)
d=np.arange(6,12).reshape(2,3)
print(c)
print(d)
```

```python
# concatenate
print(np.concatenate((c,d),axis=0)) # row concatenate 0
```

```python
# column concatenate
print(np.concatenate((c,d),axis=1))
```

EXPAND DIMENSION NP.EXPAND_DIMS

```python
# 1d to 2d to 3d
a=np.arange(15)
print(a)
print(a.shape)

# 1d to 2d
b=np.expand_dims(a,axis=0)
print(b)  #1d converted to 2d
print(b.shape)  #(1,15) 1 row 15 columns

# axis =1
c=np.expand_dims(a,axis=1)
print(c)  #(15,1) 15 row 1 column


# check shape
print(c.shape)



# this method is for column vector and row vector

```
