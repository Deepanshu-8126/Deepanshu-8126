```python

# PANDAS
import numpy as np
import pandas as pd
```

```python
# series from lists

# string
country = ['India','Pakistan','USA','Nepal','Srilanka']

pd.Series(country)
# in series a index and value here india assign the value india -0
# dtype = object means this is a string
# series there are two things values and index
```

```python
# integers ( we want to store batsman data this is in the form of integers)
runs=[13,24,56,78,100]

# we want to store this is in series
print(pd.Series(runs))

# dtype = int64
```

```python
# CUSTOM INDEX (we can provide index)
marks=[67,57,89,100]
subjects=['maths','english','science','social science']

pd.Series(marks,index=subjects)
```

```python
# setting a name (Means we can give a name to series object)

marks=pd.Series(marks,index=subjects,name='Deepanshu Marks')
print(marks)
```

```python
# Series from dictionary

calories={"day":420,"day2":380,"day3":390}
my_var=pd.Series(calories) # pass dictionary
print(my_var)
print(my_var.loc["day2"])
print(my_var.loc["day3"])




```

```python
# example--
marks={"deepanshu":56,"Himanshu":56,"vishal":65}
marks_=pd.Series(marks,name="Student marks")
print(marks_)
```

Series Attributes

```python
# size -- total items
marks_.size
```

```python
# dtype (typee of data in our series)
marks_.dtype
```

```python
# 3-- name
marks_.name
# output -- student marks

```

```python
# is_unique (repeat or not return true )
marks_.is_unique
print(marks_.is_unique)

# create a series
series=pd.Series([1,1,2,2,3,4,5,6])
print(series.is_unique) # false duplicates
```

```python
# 4--index ( return index ( index object ))
marks_.index

print(series.index) # return range index
```

```python
# 5 - values ( return values all in our series as an numpy array)
print(marks_.values)
print(series.values)
```
